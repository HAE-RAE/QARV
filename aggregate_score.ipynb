```python
import os
import pandas as pd
# df = pd.read_csv('exp/qwen1-5_05B_EN.csv')
```

```python
output = []
for fp in os.listdir('exp'):
    if fp in ['report_example.csv','.ipynb_checkpoints']:
        continue
    df = pd.read_csv(f'exp/{fp}')
    lang = fp.split('_')[-1].split('.')[0]
    model = df.model_ckpt.values[0]
    en = df.loc[:3].accuracy.mean()
    ko = df.loc[4:].accuracy.mean()
    output.append([model,lang,en,ko,df.accuracy.mean()])
```

```python
pd.DataFrame(output)
```
